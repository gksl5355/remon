```python
import json

with open("data/sample_regulations.json", "r", encoding="utf-8") as f:
    regs = json.load(f)
with open("data/sample_products.json", "r", encoding="utf-8") as f:
    prods = json.load(f)

print(regs[0])
print(prods[0])
```

```python
from glossary.term_glossary import TECH_GLOSSARY

def apply_glossary(text: str):
    for k, v in TECH_GLOSSARY.items():
        text = text.replace(k, v)
    return text

test_text = "리튬이온 전지는 3000mAh 이하의 용량을 유지해야 한다."
print(apply_glossary(test_text))
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("BAAI/bge-m3")

text = apply_glossary("흡연경고문구는 제품 포장의 30% 이상이어야 한다.")
dense_vec = model.encode(text).tolist()
sparse_vec = {w: 1.0 for w in text.split() if len(w) > 1}

print("Dense length:", len(dense_vec))
print("Sparse sample:", list(sparse_vec.items())[:5])
```

```python
from app.vectorstore.vector_client import VectorClient
from app.vectorstore.vector_schema import VectorMetadata

vc = VectorClient.from_settings()

texts = [apply_glossary(r["content"]) for r in regs]
dense_vectors = [model.encode(t).tolist() for t in texts]
sparse_vectors = [{w: 1.0 for w in t.split()} for t in texts]
metas = [
    VectorMetadata(
        clause_id=r["id"],
        type="regulation",
        country=r["country"],
        category=r["category"],
    ).model_dump()
    for r in regs
]

await vc.insert(texts, dense_vectors, sparse_vectors, metas)
```
